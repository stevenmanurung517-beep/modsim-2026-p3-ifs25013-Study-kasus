```python
# ============================
# IMPORT LIBRARY
# ============================
import simpy
import random
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import pandas as pd
from dataclasses import dataclass
import statistics

print("✅ Library berhasil diimport")
```

```python
# ============================
# KONFIGURASI SIMULASI
# ============================

@dataclass
class Config:
    """Konfigurasi Sistem Piket IT Del"""
    
    # Struktur sistem
    NUM_MEJA: int = 60
    MAHASISWA_PER_MEJA: int = 3
    
    # Total petugas = 7 orang
    PETUGAS_LAUK: int = 3
    PETUGAS_ANGKUT: int = 2
    PETUGAS_NASI: int = 2
    
    # Waktu proses (dalam menit)
    MIN_LAUK: float = 0.5
    MAX_LAUK: float = 1.0
    
    MIN_ANGKUT: float = 0.33
    MAX_ANGKUT: float = 1.0
    
    MIN_NASI: float = 0.5
    MAX_NASI: float = 1.0
    
    # Jam mulai
    START_HOUR: int = 7
    START_MINUTE: int = 0
    
    RANDOM_SEED: int = 42

print("✅ Konfigurasi simulasi siap")
```

```python
# ============================
# MODEL SIMULASI SISTEM PIKET
# ============================

class SistemPiketDES:
    
    def __init__(self, config: Config):
        self.config = config
        self.env = simpy.Environment()
        
        # Total ompreng
        self.total_ompreng = config.NUM_MEJA * config.MAHASISWA_PER_MEJA
        
        # Resource tiap tahap
        self.lauk = simpy.Resource(self.env, capacity=config.PETUGAS_LAUK)
        self.angkut = simpy.Resource(self.env, capacity=config.PETUGAS_ANGKUT)
        self.nasi = simpy.Resource(self.env, capacity=config.PETUGAS_NASI)
        
        # Buffer antar tahap
        self.buffer_lauk = simpy.Store(self.env)
        self.buffer_meja = simpy.Store(self.env)
        
        self.data = []
        
        self.start_time = datetime(2024,1,1,
                                   config.START_HOUR,
                                   config.START_MINUTE)
        
        random.seed(config.RANDOM_SEED)
        np.random.seed(config.RANDOM_SEED)
    
    def waktu_ke_jam(self, waktu):
        return self.start_time + timedelta(minutes=waktu)
    
    # =====================
    # Tahap 1: Lauk
    # =====================
    def proses_lauk(self, id_ompreng):
        with self.lauk.request() as req:
            yield req
            waktu = random.uniform(self.config.MIN_LAUK,
                                   self.config.MAX_LAUK)
            yield self.env.timeout(waktu)
        
        yield self.buffer_lauk.put((id_ompreng, self.env.now))
    
    # =====================
    # Tahap 2: Angkut (Batch 4-7)
    # =====================
    def proses_angkut(self):
        while True:
            if len(self.buffer_lauk.items) == 0:
                yield self.env.timeout(0.1)
                continue
            
            batch_size = random.randint(4,7)
            batch = []
            
            for _ in range(min(batch_size, len(self.buffer_lauk.items))):
                item = yield self.buffer_lauk.get()
                batch.append(item)
            
            with self.angkut.request() as req:
                yield req
                waktu = random.uniform(self.config.MIN_ANGKUT,
                                       self.config.MAX_ANGKUT)
                yield self.env.timeout(waktu)
            
            for item in batch:
                yield self.buffer_meja.put(item)
    
    # =====================
    # Tahap 3: Nasi
    # =====================
    def proses_nasi(self):
        while True:
            if len(self.buffer_meja.items) == 0:
                yield self.env.timeout(0.1)
                continue
            
            id_ompreng, waktu_lauk_selesai = yield self.buffer_meja.get()
            
            with self.nasi.request() as req:
                yield req
                waktu = random.uniform(self.config.MIN_NASI,
                                       self.config.MAX_NASI)
                yield self.env.timeout(waktu)
            
            selesai = self.env.now
            
            self.data.append({
                "ID": id_ompreng,
                "Waktu Selesai": selesai,
                "Durasi Total": selesai,
                "Jam Selesai": self.waktu_ke_jam(selesai)
            })
    
    # =====================
    def run(self):
        for i in range(self.total_ompreng):
            self.env.process(self.proses_lauk(i))
        
        self.env.process(self.proses_angkut())
        self.env.process(self.proses_nasi())
        
        self.env.run(until=1000)
        
        return self.analyze()
    
    def analyze(self):
        df = pd.DataFrame(self.data)
        
        hasil = {
            "Total Ompreng": len(df),
            "Durasi Total (menit)": df["Waktu Selesai"].max(),
            "Jam Selesai": df["Jam Selesai"].max(),
            "Rata-rata Durasi": df["Durasi Total"].mean()
        }
        
        return hasil, df

print("✅ Model simulasi siap")
```

```python
# ============================
# MENJALANKAN SIMULASI
# ============================

config = Config()
model = SistemPiketDES(config)
results, df = model.run()

print("SISTEM PIKET IT DEL")
print("="*50)

print(f"Total Ompreng       : {results['Total Ompreng']}")
print(f"Durasi Total        : {results['Durasi Total (menit)']:.2f} menit")
print(f"Jam Selesai         : {results['Jam Selesai'].strftime('%H:%M')}")
print(f"Rata-rata Durasi    : {results['Rata-rata Durasi']:.2f} menit")
```

```python
# ============================
# VISUALISASI HASIL
# ============================

fig, axes = plt.subplots(1,2, figsize=(12,5))

# Histogram Durasi
axes[0].hist(df["Durasi Total"], bins=20, edgecolor='black')
axes[0].set_title("Distribusi Durasi Ompreng")
axes[0].set_xlabel("Durasi (menit)")
axes[0].set_ylabel("Frekuensi")

# Timeline
axes[1].scatter(df["Waktu Selesai"], df["ID"], s=10)
axes[1].set_title("Timeline Penyelesaian Ompreng")
axes[1].set_xlabel("Waktu (menit)")
axes[1].set_ylabel("ID Ompreng")

plt.tight_layout()
plt.show()
```
